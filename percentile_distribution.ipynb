```python
import pandas as pd
import numpy as np
import time
import matplotlib.pyplot as plt
%matplotlib inline
df = pd.read_csv('E:\\history5month\\SBER_RTS_Si.csv', header= 0, error_bad_lines=False, nrows=2000000)
df.dropna(inplace=True)

windows_list = [50, 100, 200, 500, 1000]
percentile_list = []

def price_moove(values):
    return (values[-1] - values[0])/values[0]
for win in windows_list:
    df['_move'] = (df['Price1'].rolling(win).apply(price_moove))
    percentile_5_left = df._move.quantile(0.3)
    percentile_5_right = df._move.quantile(0.7)
#     ax = df._move.plot.hist(bins=50, alpha=0.5)
#     plt.pause(0.05)
    percentile_list += [(percentile_5_right-percentile_5_left)/2]
                        
lines = plt.plot(windows_list, percentile_list)
l1= lines
plt.setp(lines, linestyle='-')
plt.setp(l1, linewidth=1, color='g')
plt.grid()
plt.title('sber (x1-x0)/x0')
plt.show()  

print(percentile_list)
```
